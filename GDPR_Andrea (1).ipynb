```python
date_formats_dict = {
    # Formatos largos y específicos primero para evitar falsos positivos
    'YYYYMMDDHHMMSS': [r'(?<!\d)\d{14}(?!\d)', "%Y%m%d%H%M%S"],
    'YYYY-MM-DD HH:MM:SS': [r'(?<!\d)\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}(?!\d)', "%Y-%m-%d %H:%M:%S"],
    'DD-MM-YYYY HH:MM': [r'(?<!\d)\d{2}-\d{2}-\d{4} \d{2}:\d{2}(?!\d)', "%d-%m-%Y %H:%M"],

    # Formatos de fecha de 8 dígitos
    'YYYYMMDD': [r'(?<!\d)\d{8}(?!\d)', "%Y%m%d"],
    'DD-MM-YYYY': [r'(?<!\d)\d{2}-\d{2}-\d{4}(?!\d)', "%d-%m-%Y"],
    'YYYY-DD-MM': [r'(?<!\d)\d{4}-\d{2}-\d{2}(?!\d)', "%Y-%d-%m"], # Idéntico a YYYY-MM-DD, el orden importa
    'YYYY-MM-DD': [r'(?<!\d)\d{4}-\d{2}-\d{2}(?!\d)', "%Y-%m-%d"],
    'DD_MM_YYYY': [r'(?<!\d)\d{2}_\d{2}_\d{4}(?!\d)', "%d_%m_%Y"],
    'YYYY.MM.DD': [r'(?<!\d)\d{4}\.\d{2}\.\d{2}(?!\d)', "%Y.%m.%d"],
    'YYYY/MM/DD': [r'(?<!\d)\d{4}/\d{2}/\d{2}(?!\d)', "%Y/%m/%d"],
    'DD/MM/YYYY': [r'(?<!\d)\d{2}/\d{2}/\d{4}(?!\d)', "%d/%m/%Y"],
    'MM/DD/YYYY': [r'(?<!\d)\d{2}/\d{2}/\d{4}(?!\d)', "%m/%d/%Y"], # Idéntico a DD/MM/YYYY
    'YYYY MM DD': [r'(?<!\d)\d{4} \d{2} \d{2}(?!\d)', "%Y %m %d"],
    'DD MM YYYY': [r'(?<!\d)\d{2} \d{2} \d{4}(?!\d)', "%d %m %Y"],

    # Formatos con meses en texto
    'MonDDYYYY': [r'\b[A-Za-z]{3}\d{4}(?!\d)', "%b%d%Y"], # \b es suficiente aquí
    'DDMonYY': [r'(?<!\d)\d{2}[A-Za-z]{3}\d{2}\b', "%d%b%y"],
    'MonthDDYYYY': [r'\b[A-Za-z]+ \d{4}(?!\d)', "%B %d %Y"],
    'DDMonthYYYY': [r'(?<!\d)\d{2} [A-Za-z]+\b', "%d %B %Y"],

    # Formatos ambiguos de 6 dígitos (el orden es CRÍTICO)
    'YYMMDD': [r'(?<!\d)\d{6}(?!\d)', "%y%m%d"],
    'DDMMYY': [r'(?<!\d)\d{6}(?!\d)', "%d%m%y"],
    'MMDDYY': [r'(?<!\d)\d{6}(?!\d)', "%m%d%y"],

    # Formatos de solo tiempo
    'HHMMSS': [r'(?<!\d)\d{6}(?!\d)', "%H%M%S"], # Ambiguo con fechas de 6 dígitos
    'HH:MM:SS': [r'(?<!\d)\d{2}:\d{2}:\d{2}(?!\d)', "%H:%M:%S"],
}
```

```python
import re
import datetime

def extract_data_from_filenames(list_of_filenames: list) -> list:
    """
    Recorre una lista de nombres de archivo para extraer fechas y dígitos finales.

    Args:
        list_of_filenames: Una lista de strings con nombres de archivo.

    Returns:
        Una lista de diccionarios, cada uno con el nombre del archivo original,
        la fecha encontrada (como objeto datetime), la cadena de fecha original
        y los dígitos extraídos.
    """
    # El diccionario con las expresiones regulares mejoradas
    date_formats_dict = {
        'YYYYMMDDHHMMSS': [r'(?<!\d)\d{14}(?!\d)', "%Y%m%d%H%M%S"],
        'YYYY-MM-DD HH:MM:SS': [r'(?<!\d)\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}(?!\d)', "%Y-%m-%d %H:%M:%S"],
        'DD-MM-YYYY HH:MM': [r'(?<!\d)\d{2}-\d{2}-\d{4} \d{2}:\d{2}(?!\d)', "%d-%m-%Y %H:%M"],
        'YYYYMMDD': [r'(?<!\d)\d{8}(?!\d)', "%Y%m%d"],
        'DD-MM-YYYY': [r'(?<!\d)\d{2}-\d{2}-\d{4}(?!\d)', "%d-%m-%Y"],
        'YYYY-MM-DD': [r'(?<!\d)\d{4}-\d{2}-\d{2}(?!\d)', "%Y-%m-%d"],
        'DD_MM_YYYY': [r'(?<!\d)\d{2}_\d{2}_\d{4}(?!\d)', "%d_%m_%Y"],
        'YYYY.MM.DD': [r'(?<!\d)\d{4}\.\d{2}\.\d{2}(?!\d)', "%Y.%m.%d"],
        'YYYY/MM/DD': [r'(?<!\d)\d{4}/\d{2}/\d{2}(?!\d)', "%Y/%m/%d"],
        'DD/MM/YYYY': [r'(?<!\d)\d{2}/\d{2}/\d{4}(?!\d)', "%d/%m/%Y"],
        'YYYY MM DD': [r'(?<!\d)\d{4} \d{2} \d{2}(?!\d)', "%Y %m %d"],
        'DDMonthYYYY': [r'(?<!\d)\d{2} [A-Za-z]+ \d{4}\b', "%d %B %Y"],
        'YYMMDD': [r'(?<!\d)\d{6}(?!\d)', "%y%m%d"],
        'HHMMSS': [r'(?<!\d)\d{6}(?!\d)', "%H%M%S"],
        'HH:MM:SS': [r'(?<!\d)\d{2}:\d{2}:\d{2}(?!\d)', "%H:%M:%S"],
    }

    all_results = []

    for filename in list_of_filenames:
        extracted_date = None
        original_date_string = None
        extracted_digits = None

        # --- 1. Proceso de extracción de fecha ---
        # Itera sobre los formatos para encontrar una fecha válida
        for format_name, (pattern, strp_format) in date_formats_dict.items():
            match = re.search(pattern, filename)
            if match:
                date_str = match.group(0)
                try:
                    # Intenta validar que la cadena encontrada es una fecha real
                    extracted_date = datetime.datetime.strptime(date_str, strp_format)
                    original_date_string = date_str
                    # Si tiene éxito, hemos encontrado la fecha, sal del bucle
                    break
                except ValueError:
                    # La regex coincidió, pero no es una fecha válida (ej: 32/13/2024)
                    # Continúa al siguiente formato
                    continue

        # --- 2. Proceso de extracción de dígitos ---
        # Busca un patrón de la forma "_12345.csv" al final del archivo
        digit_match = re.search(r'_(\d+)\.csv$', filename)
        if digit_match:
            extracted_digits = digit_match.group(1)

        # --- 3. Guardar resultados ---
        all_results.append({
            'filename': filename,
            'fecha_extraida': extracted_date,
            'string_fecha_encontrado': original_date_string,
            'digitos_extraidos': extracted_digits
        })

    return all_results

### Ejemplo de Uso

# Muestra de la lista de archivos generada anteriormente
sample_filenames = [
    'AR_20240611_934.csv',
    'MX_20240611_5103789.csv',
    'PE_20240611_19934567.csv',
    'DE_11-06-2024_918273.csv',
    'ES_11_06_2024_8273.csv',
    'FR_2024.06.11_192837.csv',
    'GB_2024/06/11_82731.csv',
    'IT_11 06 2024_82731.csv',
    'JP_June 11 2024_8273.csv',
    'KR_11Jun24_5827.csv',
    'VE_20240611153045_8273.csv', # Fecha con hora
    'NG_15:30:45_98273.csv' # Solo hora
]

# Ejecutar la función
results = extract_data_from_filenames(sample_filenames)

# Imprimir los resultados de forma clara
for res in results:
    print(f"Archivo: {res['filename']}")

    if res['fecha_extraida']:
        # Formato
        formatted_date = res['fecha_extraida'].strftime('%Y-%m-%d %H:%M:%S')
        print(f"  -> Fecha encontrada: '{res['string_fecha_encontrado']}' -> Validada como: {formatted_date}")
    else:
        print(f"  -> Fecha encontrada: NINGUNA")

    print(f"  -> Dígitos encontrados: {res['digitos_extraidos'] or 'NINGUNO'}")
    print("-" * 30)
```
